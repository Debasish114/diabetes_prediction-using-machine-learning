# outlier detection

```python
from collections import Counter
def outlier_detection(df,features):
    outlier_indices=[]
    for c in features:
        q1=df[c].quantile(0.25)
        q3=df[c].quantile(0.75)
        iqr=q3-q1
        ub=q3+(1.5*iqr)
        lb=q1-(1.5*iqr)
        #detect otlier and its indices
        outlier_list=df[(df[c]<lb) | (df[c]>ub)].index
        #append the found outlier indices for col to the list of outlier_indices
        outlier_indices.extend(outlier_list)
    # select observations containing more than 2 outliers
    outlier_indices=Counter(outlier_indices)
    multiple_outliers=list(k for k,v in outlier_indices.items() if v>2)
    return multiple_outliers
```

```python
#displaying the outliers
df.loc[outlier_detection(df,["age","bmi","HbA1c_level","blood_glucose_level"])]
```

```python
#dropping the outliers
df=df.drop(outlier_detection(df,["age","bmi","HbA1c_level","blood_glucose_level"]),axis=0).reset_index()
```
